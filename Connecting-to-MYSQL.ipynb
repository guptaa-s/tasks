```python
import pandas as pd
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00448/carbon_nanotubes.csv",sep=';')
df
```

```python
import mysql.connector as connection
def show_databases():
    try:
        mydb=connection.connect(host="localhost", user="root", passwd="", use_pure=True)
        query="SHOW DATABASES"
        cursor=mydb.cursor()
        cursor.execute(query)
        print(cursor.fetchall())
    except Exception as e:
        mydb.close()
        print(str(e))

def create_database(databasename):
    try:
        mydb=connection.connect(host="localhost", user="root", passwd="", use_pure=True)
        query="CREATE DATABASE {}".format(databasename)
        cursor=mydb.cursor()
        cursor.execute(query)
        print("Database created")
    except Exception as e:
        mydb.close()
        print(str(e))

def create_table(databasename,tablename,var):
    try:
        mydb=connection.connect(host="localhost",database=databasename, user="root", passwd="", use_pure=True)
        query="CREATE TABLE {} {}".format(tablename,var)
        cursor=mydb.cursor()
        cursor.execute(query)
        print("Table created")
    except Exception as e:
        mydb.close()
        print(str(e))
        
def insert_values(databasename,tablename,val):
    try:
        mydb=connection.connect(host="localhost",database=databasename, user="root", passwd="", use_pure=True)
        query="INSERT INTO {} VALUES {}".format(tablename,val)
        cursor=mydb.cursor()
        cursor.execute(query)
        mydb.commit()
        print("VALUE INSERTED")
    except Exception as e:
        mydb.close()
        print(str(e))


    
```

```python
show_databases()
```

```python
create_database("carbon")
```
